### 고성능 연산

```
nrows, ncols = 100000, 100
df1, df2,df3 ,df4 = (pd.DataFrame(np.random.rand(nrows, ncols)) for i in range(4))
```

```
df1.shape
```

```
%timeit df1 + df2 + df3 + df4
```

```
%timeit pd.eval('df1 + df2 + df3 + df4') # 고성능연산은 속도가빠르다
```

```
%timeit df1 * -df2 / (-df3 * df4)
```

```
%timeit pd.eval('df1 * -df2 / (-df3 * df4)')
```

```
%timeit (df1 < df2) & (df2 <= df3) & (df3 != df4)
```

```
%timeit pd.eval('(df1 < df2) & (df2 <= df3) & (df3 != df4)')
```

```
df = pd.DataFrame(np.random.rand(1000000, 5), columns = ['A','B','C','D','E'])
df.head()
```

```
%timeit df['A'] + df['B'] / df['C'] - df['D'] * df['E']
```

```
%timeit pd.eval('df.A + df.B / df.C - df.D * df.E')
```

```
%timeit df.eval('A + B / C - D * E') # column명으로 접근해서 계산도 가능
```

```
df.eval('R = A + B / C - D * E', inplace = True) # 계산결과 바로반영도 가능
df.head()
```

```
df.eval('R = A - B / C * D * E', inplace = True) # 계산실이바꿔 업데이트도 가능
df.head()
```

```
col_mean = df.mean(1)
df['A'] + col_mean
```

```
df.eval('A + @col_mean') # @로 외부에있는 변수를 가져와 계산가능
```

```
df[(df.A < 0.5) & (df.B < 0.5) & (df.C > 0.5)]
```

```
pd.eval('df[(df.A < 0.5) & (df.B < 0.5) & (df.C > 0.5)]') # 더 빠름
```

```
# 위의 방법이 쓰는게 많아 귀찮으니 query로 질의문 형태로 접근할 수 있다.
df.query('(A < 0.5) and (B < 0.5) and (C > 0.5)') # 이러한 조건이 해당하는 행들을 출력으로 얻고싶으면 eval말고 query 쓰면된다.
# eval은 bool값 리턴한다.
```

```
col_mean = df['D'].mean()
df[(df.A < col_mean) & (df.B < col_mean)]
```

```
df.query('A < @col_mean and B < @col_mean')
```

```
def create_df(cols, idx):
  data = {c:[str(c.lower()) + str(i) for i in idx] for c in cols}
  return pd.DataFrame(data, idx)
```

### 병합과 조인

```
df1 = pd.DataFrame({'학생': ['홍길동','이순신','임꺽정','김유신'],
                    '학과': ['경영학과','교육학과','컴퓨터학과','통계학과']})
df1
```

```
df2 = pd.DataFrame({'학생': ['홍길동','이순신','임꺽정','김유신'],
                    '입학년도' : [2012,2016,2019,2020]})
df2
```

```
df3 = pd.merge(df1, df2) # 병합, concat이 합치는것이라면 병합은 열, 행 들이 중복되지않게 합친다
df3
```

```
df4 = pd.DataFrame({'학과' : ['경영학과', '교육학과','컴퓨터학과','통계학과'],
                    '학과장' : ['황희','장영실','안창호','정약용']})
df4
```

```
pd.merge(df3, df4)
```

```
df5 = pd.DataFrame({'학과' : ['경영학과','교육학과','교육학과','컴퓨터학과','컴퓨터학과','통계학과'],
                    '과목' : ['경영개론','기초수학','물리학','프로그래밍','운영체제','확률론']})
df5
```

```
pd.merge(df1, df5) # 기준이되는 학과를 바탕아르 데이터를 만들다보니 학생이 중복된다.
```

```
pd.merge(df1, df2, on='학생') # 기준을 on으로 정할 수 있다. 둘다있는 열로 해야한다.
```

```
df6 = pd.DataFrame({'이름': ['홍길동','이순신','임꺽정','김유신'],
                    '성적': ['A','A+','B','A+']})
df6
```

```
df1
```

```
pd.merge(df1, df6 , left_on= '학생', right_on='이름') # 중복되는 열이 없을경우 left_on, right_on으로 각각 기준 정해줘야한다
```

```
pd.merge(df1,df6, left_on='학생', right_on='이름').drop('이름',axis=1) # drop으로 중복되는 열 지움
```

```
mdf1 = df1.set_index('학생')
mdf2 = df2.set_index('학생')
```

```
mdf1
```

```
mdf2
```

```
pd.merge(mdf1, mdf2 , left_index=True, right_index=True) # index를 지정
# pd.merge(mdf1, mdf2) # 그냥하면 지정하라고뜸 :  No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False
```

```
mdf1.join(mdf2) # 인덱스를 기준으로 합침
```

```
# left_on, right_on은 기준으로 잡을 열을 정할때 사용
# left_index, right_index는 index를 기준으로 잡아야할때 사용
# 둘다 혼용해서 사용해도 된다
pd.merge(mdf1, df6, left_index=True, right_on='이름')
```
