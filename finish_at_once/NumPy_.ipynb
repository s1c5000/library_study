### 배열 속성 정보

```
def array_info(array):
  print(array)
  print('ndim:', array.ndim)
  print('shape:' , array.shape)
  print('dtype:', array.dtype)
  print('size:', array.size) # 요소개수
  print('itemsize:', array.itemsize) # item의 크기 단위 byte
  print('nbytes:', array.nbytes) # 배열의 크기 단위 byte
  print('strides:', array.strides) # 다음차원으로 넘어가기위해 필요한 byte, 1차원이면 itemsize와 같다
```

## 배열 값 삽입/수정/삭제/복사

### 배열 값 삽입

* `insert()`: 배열의 특정 위치에 값 삽입
* axis를 지정하지 않으면 1차원 배열로 변환
* 추가할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)
b1 = np.insert(a1, 0, 10)
print(b1)
c1 = np.insert(a1, 2, 10)
print(c1)
```

```
print(a2)
b2 = np.insert(a2, 1, 10, axis = 0)
print(b2)
c2 = np.insert(a2, 1, 10, axis = 1)
print(c2)
```

### 배열 값 수정

* 배열의 인덱싱으로 접근하여 값 수정

```
print(a1)
a1[0] = 2
a1[2] = 3
print(a1)
a1[:3] = 9
print(a1)
i = np.array([1,3,4])
a1[i] = 0
print(a1)
a1[i] += 4
print(a1)
```

```
print(a2)
a2[0,0] = 1
a2[1,1] = 2
a2[2] = 3
print(a2)
print()
a2[1:, 2] = 9
print(a2)
row = [0,1]
col = [1,2]
a2[row, col] = 0
print(a2)
```

### 배열 값 삭제

* `delete()`: 배열의 특정 위치에 값 삭제
* axis를 지정하지 않으면 1차원 배열로 변환
* 삭제할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)
b1 = np.delete(a1,1) # deep copy된다
print(b1) 
print(a1)
```

```
print(a2)
b2 = np.delete(a2,1, axis = 0)
print(b2)
c2 = np.delete(a2,0, axis = 1)
print(c2)
```

### 배열 복사

* 리스트 자료형과 달리 배열의 슬라이스는 복사본이 아님

```
print(a2)
print(a2[:2, :2])
a2_sub = a2[:2, :2] # numpy는 슬라이스시 주소를 반환
print(a2_sub)
a2_sub[:, 1] = 0
print(a2_sub)
print(a2) # 원본배열 바뀜
```


* `copy()`: 배열이나 하위 배열 내의 값을 명시적으로 복사

```
print(a2)
a2_sub_copy = a2[:2, :2].copy() # deep copy, 주소다름
print(a2_sub_copy)
a2_sub_copy[:,1] = 1
print(a2_sub_copy)
print(a2) # 원본 바뀌지않음
```

### 배열 분할

* `split()`: 배열 분할

```
a1 = np.arange(10,20)
print(a1)
b1, c1 = np.split(a1,[5]) # 인덱스위치로 나눔
print(b1, c1)
b1, b2, b3, b4, b5 = np.split(a1,[2,4,6,8])
print(b1,b2,b3,b4, b5)
print()
a1 = np.arange(1,10).reshape(3,3)
b1, c1 = np.split(a1,[1]) # 가장 큰 단위를 자름
print(b1)
print(c1)
```

* `vsplit()`: 수직 분할, 1차원으로 분할

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)
b2, c2 = np.vsplit(a2, [2]) # 가장 큰 단위를 자름
print(b2)
print(c2)
print(b2.shape, c2.shape)
```

* `hsplit()`: 수평 분할, 2차원으로 분할

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)
b2, c2 = np.vsplit(a2, [2]) # # 첫번재로 큰 단위 자름
print(b2)
print(c2)
print()

b2, c2 = np.hsplit(a2, [2]) # 두번째로 큰 단위를 자름
print(b2)
print(c2)
print(b2.shape, c2.shape)
```

* `dsplit()`: 깊이 분할, 3차원으로 분할

```
a3 = np.arange(1,37).reshape(3,4,3)
print(a3)
print('vsplit ###########################')
b3, c3 = np.vsplit(a3, [2]) # 첫번재로 큰 단위 자름
print(b3)
print(c3)
print(b3.shape, c3.shape)
print('hsplit ###########################')
b3, c3 = np.hsplit(a3, [2]) # 두번재로 큰 단위 자름
print(b3)
print(c3)
print(b3.shape, c3.shape)
print('dsplit ###########################')
b3, c3 = np.dsplit(a3, [2]) # 3번째로 큰 단위를 자름
print(b3)
print(c3)
print(b3.shape, c3.shape)
```

```
a3 = np.arange(1,49).reshape(2,2,3,4)
print(a3)
print('vsplit ###########################')
b3, c3 = np.vsplit(a3, [1]) # 첫번재로 큰 단위 자름
print(b3)
print(c3)
print(b3.shape, c3.shape)
print('hsplit ###########################')
b3, c3 = np.hsplit(a3, [1]) # 두번재로 큰 단위 자름
print(b3)
print(c3)
print(b3.shape, c3.shape)
print('dsplit ###########################')
b3, c3 = np.dsplit(a3, [1]) # 3번째로 큰 단위를 자름
print(b3)
print(c3)
print(b3.shape, c3.shape)
```

## 배열 연산

* NumPy의 배열 연산은 벡터화(vectorized) 연산을 사용
* 일반적으로 NumPy의 범용 함수(universal functions)를 통해 구현
* 배열 요소에 대한 반복적인 계산을 효율적으로 수행     
image.png

### 브로드캐스팅(Broadcasting)

```
a1 = np.array([1,2,3])
print(a1)
print(a1 + 5)
print('#' * 20)
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print(a2 + a1)
print('#' * 20)
b2 = np.array([1,2,3]).reshape(3,1)
print(a1 + b2)
```

### 산술 연산(Arithmetic Operators)    
image.png

#### 불리언 연산자(Boolean Operators)
image.png

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print((a2 > 5) & (a2 < 8)) # and
print(a2[(a2 > 5) & (a2 < 8)]) # 조건으로도 사용가능

print((a2 > 5) | (a2 < 8)) # or
print(a2[(a2 > 5) |(a2 < 8)]) # 조건으로도 사용가능

print((a2 > 5) ^ (a2 < 8)) # xor
print(a2[(a2 > 5) ^ (a2 < 8)]) # 조건으로도 사용가능

print(~(a2 > 5)) # not
print(a2[(~a2 > 5)]) # 조건으로도 사용가능
```
